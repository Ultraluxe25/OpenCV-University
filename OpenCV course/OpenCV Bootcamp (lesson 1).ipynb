```python
# Download resources
import os
from zipfile import ZipFile
from urllib.request import urlretrieve


def download_and_unzip(url, save_path):
    print(f'Downloading and extracting resources...')
    
    # Downloading zip file using urllib package
    urlretrieve(url, save_path)
    
    try:
        # Extracting zip file using zipfile package
        with ZipFile(save_path) as z:
            # Extracting zip file in the same directory
            z.extractall(os.path.split(save_path)[0])
        print("Done")
    except Exception as e:
        print('Error', e)
```

```python
# Save resources
url = r"https://www.dropbox.com/s/qhhlqcica1nvtaw/opencv_bootcamp_assets_NB1.zip?dl=1"
my_path = os.path.join(os.getcwd() + '\\resources\\lesson 1', f"opencv_bootcamp_assets_NB1.zip")

# Unzip resources 
if not os.path.exists(my_path):
    download_and_unzip(url, my_path)
```

```python
lake = 'resources/lesson 1/New_Zealand_Lake.jpg'
lake_img = cv2.imread(lake)
b, g, r = cv2.split(lake_img)  # split each layer

# Show the channels
plt.figure(figsize=(20, 5))

# Blue channel
plt.subplot(141)
plt.imshow(X=b, cmap='gray')
plt.title(label='Blue Channel')

# Green channel
plt.subplot(142)
plt.imshow(X=g, cmap='gray')
plt.title(label='Green Channel')

# Blue channel
plt.subplot(143)
plt.imshow(X=r, cmap='gray')
plt.title(label='Red Channel')

# Merged channel
plt.subplot(144)
plt.imshow(X=lake_img[:, :, ::-1])  # Switch from BGR to RGB
plt.title(label='Merged Channel');
```

```python
# Convert from BGR to RGB
rgb_lake = cv2.cvtColor(src=lake_img, code=cv2.COLOR_BGR2RGB)
plt.imshow(X=rgb_lake);
```

```python
# Convert to HSV color space
hsv_lake = cv2.cvtColor(src=lake_img, code=cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(m=hsv_lake)
plt.figure(figsize=(20, 5))

# Hue channel
plt.subplot(141)
plt.imshow(X=h, cmap='gray')
plt.title(label='H Channel')

# Saturation channel
plt.subplot(142)
plt.imshow(X=s, cmap='gray')
plt.title(label='S Channel')

# Value channel
plt.subplot(143)
plt.imshow(X=v, cmap='gray')
plt.title(label='V Channel')

# Merged channel
plt.subplot(144)
plt.imshow(X=hsv_lake)  # Switch from BGR to RGB
plt.title(label='HSV Channel');
```

### HSV (англ. Hue, Saturation, Value — тон, насыщенность, значение)

```python
# Modify each parameter in HSV
# Convert to HSV color space
hsv_lake = cv2.cvtColor(src=lake_img, code=cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(m=hsv_lake)
plt.figure(figsize=(20, 5))

# Hue channel
plt.subplot(141)
plt.imshow(X=h + 150, cmap='gray')
plt.title(label='H+150 Channel')

# Saturation channel
plt.subplot(142)
plt.imshow(X=s + 100, cmap='gray')
plt.title(label='S+100 Channel')

# Value channel
plt.subplot(143)
plt.imshow(X=v + 50, cmap='gray')
plt.title(label='V+50 Channel')

# Merged channel
plt.subplot(144)
plt.imshow(X=hsv_lake)  # Switch from BGR to RGB
plt.title(label='HSV Channel');
```
