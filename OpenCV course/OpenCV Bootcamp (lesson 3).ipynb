# Image Annotation

```python
# Download and unzip data
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from zipfile import ZipFile
from urllib.request import urlretrieve
from IPython.display import Image


def download_and_unzip(url, save_path):
    print('Downloading...')
    urlretrieve(url, save_path)
    try:
        with ZipFile(save_path) as z:
            # Extract into same directory
            z.extractall(os.path.split(save_path)[0])
        print('Done')
    except Exception as e:
        print(f'Exception: {e}')


web = r"https://www.dropbox.com/s/48hboi1m4crv1tl/opencv_bootcamp_assets_NB3.zip?dl=1"
save_folder = os.path.join(os.getcwd(), 'resources', 'lesson 3')
save_into = os.path.join(save_folder, 'opencv_bootcamp_assets_NB3.zip')

# Create folder 'lesson x if not exists'
if not os.path.exists(save_folder):
    os.makedirs(name=save_folder)
    
if not os.path.exists(save_into):
    download_and_unzip(web, save_into)
```

```python
image_path = os.path.join(os.getcwd(), 'resources', 'lesson 3', 'Apollo_11_Launch.jpg')

img = cv2.imread(filename=image_path, flags=cv2.IMREAD_COLOR)
plt.imshow(X=img[:, :, ::-1]);
```

```python
# Draw the line
img_line = img.copy()

cv2.line(img=img_line, pt1=(300, 300), pt2=(800, 300), color=(0, 255, 255), thickness=10, lineType=cv2.LINE_AA)
plt.imshow(X=img_line[:, :, ::-1]);
```

```python
# Draw the circle
img_circle = img.copy()

cv2.circle(img=img_circle, center=(650, 300), radius=100, color=(255, 0, 255), thickness=10, lineType=cv2.LINE_AA)
plt.imshow(X=img_circle[:, :, ::-1]);
```

```python
# Draw the rectangle
img_rectangle = img.copy()

cv2.rectangle(img=img_rectangle, pt1=(500, 100), pt2=(700, 600), color=(255, 255, 0), thickness=10, lineType=cv2.LINE_AA)
plt.imshow(X=img_rectangle[:, :, ::-1]);
```

```python
# Put text
img_text = img.copy()
text = 'Apollo 11 Saturn V Launch, July 16, 1969'
fontFace = cv2.FONT_HERSHEY_SIMPLEX
fontScale = 1.8
color = (0, 0, 255)
thickness = 3

cv2.putText(img_text, text, (50, 700), fontFace, fontScale, color, thickness)
plt.imshow(img_text[:, :, ::-1]);
```
