```python
import os
from urllib.request import urlretrieve
from zipfile import ZipFile
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
# Downloading and extracting
def download_and_unzip(url, save_path):
    print(f"Downloading and extracting assests....", end="")

    # Downloading zip file using urllib package.
    urlretrieve(url, save_path)

    try:
        # Extracting zip file using the zipfile package.
        with ZipFile(save_path) as z:
            # Extract ZIP file contents in the same directory.
            z.extractall(os.path.split(save_path)[0])

        print("Done")

    except Exception as e:
        print("\nInvalid file.", e)
        
        
URL = r"https://www.dropbox.com/s/rys6f1vprily2bg/opencv_bootcamp_assets_NB2.zip?dl=1"

asset_zip_path = os.path.join(os.getcwd() + '\\resources\\lesson 2', f"opencv_bootcamp_assets_NB2.zip")

# Download if assest ZIP does not exists. 
if not os.path.exists(asset_zip_path):
    download_and_unzip(URL, asset_zip_path)  
```

```python
img = cv2.imread(filename=path + 'checkerboard_18x18.png', flags=0)
plt.imshow(X=img, cmap='gray')
print(img)
```

```python
ship = cv2.imread(filename=path + 'New_Zealand_Boat.jpg', flags=cv2.IMREAD_COLOR)
ship = ship[:, :, ::-1]
plt.imshow(ship);
```

```python
ship.shape
```

```python
# Cropped image
cropped_ship = ship[200:400, 300:600]
print(cropped_ship.shape[:2])
plt.imshow(X=cropped_ship);
```

```python
# Resize image (In 2 times bigger)
resized_img = cv2.resize(src=cropped_ship, dsize=None, fx=2, fy=2)
print(resized_img.shape[:2])
plt.imshow(X=resized_img);
```

# [cv2.resize()](https://www.tutorialkart.com/opencv/python/opencv-python-resize-image/#gsc.tab=0)

```python
# Resize image with special size
width = 100
height = 200
dim = (width, height)

resized = cv2.resize(src=cropped_ship, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(X=resized);
```

```python
# Save the image
cv2.imwrite(filename=path + 'resized_cropped_ship.png', img=cropped_ship[:, :, ::-1])
```

```python
# Flip images
img_hors = cv2.flip(src=ship, flipCode=1)
img_vert = cv2.flip(src=ship, flipCode=0)
img_flipped_both = cv2.flip(src=ship, flipCode=-1)

# Show images
plt.figure(figsize=(20, 5))
plt.subplot(141)
plt.imshow(X=img_hors)
plt.title('Horizontal Flip')

plt.subplot(142)
plt.imshow(X=img_vert)
plt.title('Vertical Flip')

plt.subplot(143)
plt.imshow(X=img_flipped_both)
plt.title('Both Flip')

plt.subplot(144)
plt.imshow(X=ship)
plt.title('Original ship');
```
